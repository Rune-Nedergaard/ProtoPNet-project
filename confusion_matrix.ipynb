```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
import ast


import seaborn as sns
```

```python
cm = pd.read_csv('cm.csv')
cm
```

```python
cm = cm.drop('Unnamed: 0', axis=1)
cm
```

```python
labels = ['Fetal abdomen', 'Fetal femur, Fetal thorax, Maternal cervix, Trans-cerebellum, Trans-thalmic, Trans-ventricular']
#set the labels as index
cm.set_axis(labels, axis=1,inplace=True)
cm
```

```python
#we plot the confusion matrix
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, cmap="Blues", fmt='g')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
#show legend outside the plot
#make labels the legend
labels = {0: 'Fetal abdomen', 1: 'Fetal femur', 2: 'Fetal thorax', 3: 'Maternal cervix',4: 'Trans-cerebellum',5: 'Trans-thalmic', 6: 'Trans-ventricular'}
plt.legend(labels=labels, loc='center left', bbox_to_anchor=(1, 0.5))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python

g = sns.heatmap(cm, annot=True,  fmt= 'g',cmap='Reds') #annot=True to annotate cells
#specify figure size

g.figure.set_size_inches(10,6)
#set the size of the values in the heatmap
#set size of annotations
#set xlabel and ylabel to arial
g.set_xlabel('Predicted labels', fontname = 'Arial', fontsize = 18)
g.set_ylabel('True labels', fontname = 'Arial', fontsize = 18)
#make font of labels bold
g.xaxis.label.set_fontweight('normal')
g.yaxis.label.set_fontweight('normal')
#make xticks in middle of the cell
#g.set_xticks(range(len(labels)), labels=labels, rotation=45, horizontalalignment='right')
g.set_yticks(range(len(labels)), labels=labels, rotation=0, horizontalalignment='right')
#set font to arial
g.set_xticklabels(labels, fontname='Arial', fontsize=15)
g.set_yticklabels(labels, fontname='Arial', fontsize=15)
#also for title
#g.set_title(label = 'Confusion Matrix', fontname = 'Arial', fontsize=20, fontweight='normal')

#g.set_xticklabels(['Fetal abdomen', 'Fetal femur', 'Fetal thorax', 'Maternal cervix', 'Trans-cerebellum', 'Trans-thalmic', 'Trans-ventricular']) 
g.set_yticklabels(['Fetal abdomen (0)', 'Fetal femur (1)', 'Fetal thorax (2)', 'Maternal cervix (3)', 'Trans-cerebellum (4)', 'Trans-thalmic (5)', 'Trans-ventricular (6)'])

#draw a box around the last three rows and columns
g.add_patch(plt.Rectangle((4, 4), 2.99, 2.99, fill=False, edgecolor='black', lw=2))
plt.show()
```
